```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tpu = pd.read_csv('TPU.csv')
gpu = pd.read_csv('GPU.csv')
cpu = pd.read_csv('CPU.csv')
```

```python
df = pd.concat([cpu,gpu,tpu]).reset_index().drop('index',axis=1)
```

```python
df
```

```python
sns.catplot(x='Processing Unit',y='Epoch 10',hue='Batch Size',data=df)
```

```python
df.plot(x='Processing Unit', y=["Epoch 1", "Epoch 5", "Epoch 10"], kind="bar")
```

```python
dft = df[['Processing Unit','Batch Size','Epoch 10']]
```

```python
dft_cpu = dft[dft['Processing Unit']=='CPU']
dft_gpu = dft[dft['Processing Unit']=='GPU']
dft_tpu = dft[dft['Processing Unit']=='TPU']
```

```python
plt.plot(dft_cpu['Batch Size'],dft_cpu['Epoch 10'], color='r', label='CPU')
plt.plot(dft_gpu['Batch Size'],dft_gpu['Epoch 10'],color='g', label='GPU')
plt.plot(dft_tpu['Batch Size'],dft_tpu['Epoch 10'],color='b', label='TPU')
plt.xlabel('Batch Size')
plt.ylabel('Total Time(seconds)')
plt.legend()
```

```python
plt.plot(dft_gpu['Batch Size'],dft_gpu['Epoch 10'],color='g', label='GPU')
plt.plot(dft_tpu['Batch Size'],dft_tpu['Epoch 10'],color='b', label='TPU')
plt.xlabel('Batch Size')
plt.ylabel('Total Time(seconds)')
plt.legend()
```
