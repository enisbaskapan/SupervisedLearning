```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, datasets, optimizers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, accuracy_score
import tensorflow_hub as hub
import cv2
```

```python
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
plt.imshow(X_train[0])
```

```python
cnn = models.Sequential([

    #cnn
    layers.Conv2D(filters=32, kernel_size=(3,3), input_shape=(32,32,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    #dense
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  
])

cnn.compile(optimizer = optimizers.Adam(learning_rate=0.001),
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
             )
result = cnn.fit(X_train, y_train, epochs=10, batch_size=500, validation_data=(X_test, y_test))
```

```python
cnn.evaluate(X_test, y_test)
```

```python
plt.figure(figsize=(15,8))
plt.plot(result.history['loss'], color='teal', label='loss')
plt.plot(result.history['val_loss'], color='orange', label='val_loss')
plt.legend()
```

# Data Augmentation

```python
#Used to alter existing images to create new ones to solve overfitting
```

```python
data_augmentation = models.Sequential([
    layers.RandomFlip(input_shape=(32,32,3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.2)
])
```

```python
aug_cnn = models.Sequential([
    #data augmentation
    data_augmentation,
    #cnn
    layers.Conv2D(filters=32, kernel_size=(3,3), input_shape=(32,32,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    #dropout
    layers.Dropout(0.2),
    #dense
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')  
])

aug_cnn.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
             )
aug_result = aug_cnn.fit(X_train, y_train, epochs=10, batch_size=500, validation_data=(X_test, y_test))
```

```python
aug_cnn.evaluate(X_test, y_test)
```
