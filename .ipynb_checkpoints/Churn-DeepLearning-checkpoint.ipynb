```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('./data/Churn_Modelling.csv')
df.head()
```

```python
df.drop(['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
tenure_exit = df[df['Exited']==1].Tenure
tenure_stay = df[df['Exited']==0].Tenure

plt.hist([tenure_exit, tenure_stay], label=['Exited','Stayed']);
plt.xlabel('Tenure')
plt.ylabel('Number of Customers')
plt.legend()
```

```python
X = df.drop('Exited',axis=1)
y = df.Exited
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
len(X_train.columns)
```

```python
model = keras.Sequential([
    layers.Dense(7, input_shape=(11,), activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')
    
])

model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )
result = model.fit(X_train_scaled, y_train, epochs=50, validation_data=(X_test_scaled, y_test))
```

```python
model.evaluate(X_test_scaled, y_test)
```

```python
pred_array = model.predict(X_test_scaled)
```

```python
y_pred = []
for pred in pred_array:
    if pred < 0.4:
        y_pred.append(0)
    else:
        y_pred.append(1)
```

```python
pd.DataFrame(list(zip(y_test,y_pred)), columns=['True','Pred'])
```

```python
print(classification_report(y_test, y_pred))
```

# SMOTE

```python
sm = SMOTE(random_state=0)
X_res, y_res = sm.fit_resample(X, y)
```

```python
X_res = pd.get_dummies(X_res, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = keras.Sequential([
    layers.Dense(7, input_shape=(11,), activation='relu'),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')
    
])

model.compile(optimizer = 'SGD',
              loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )
result_res = model.fit(X_train_scaled, y_train, epochs=300, batch_size=10, validation_data=(X_test_scaled, y_test))
```

```python
pred_array_res = model.predict(X_test_scaled)

y_pred_res = []
for pred in pred_array_res:
    if pred < 0.4:
        y_pred_res.append(0)
    else:
        y_pred_res.append(1)

print(classification_report(y_test, y_pred_res))
```
